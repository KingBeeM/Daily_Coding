## 사용자 정의 함수

### Docstring (문서화)
- 참조 : https://wikidocs.net/16050

```python
def hello():
  """
  This function prints
  'Hello World!'
  """
  print("Hello World!")

hello()
```

```python
help(hello())
```

```python
def hello(name: str) -> None:
  """
  This function takes parameters and returns them.
  prints "Hello" + name
  If it is not a string type, a TypeError is displayed.
  """
  if not isinstance(name, str):
    raise TypeError("Name Must be a string.")
  print(f"Hello, {name}")

hello("king")
```

```python
hello(123)
```

```python
def add(a, b, c):
  """
  This function takes a parameter, adds it, and returns it.
  But Parameters are limited to 3.
  """
  return a + b + c

print(add(1, 2, 3))
```

```python
print(add(1, 2, 3, 4)
```

### *args
- 참조 : https://wikidocs.net/24

```python
def add(*args):
  print(args, type(args))   

add(1, 2, 3)
```

```python
def add(*args):
  """
  This function takes a parameter, adds it, and returns it.
  The number of parameters is not limited.
  """
  addition = 0
  for number in args:
    addition += number
  return addition

print(add(1, 2))
print(add(1, 2, 3))
print(add(1, 2, 3, 4))
```

### **kwargs
- 머신러닝 클래스, 하이퍼 파라미터 튜닝 시 , 입력값이 딕셔너리를 기본값으로 으로 받음

```python
def temp(**kwargs):
  for key, value in kwargs.items():
    print(key, value)

temp(name="king", age=30, nationality="Korea")
```

### iloc & loc
- iloc
 +  데이터프레임의 행이나 컬럼에 인덱스 값으로 접근.
 + integer location의 약어로, 컴퓨터가 읽을 수 있는 indexing 값으로 데이터에 접근하는 것이다.

- loc
 + 데이터프레임의 행이나 컬럼에 label이나 boolean array로 접근.
 + location의 약어로, 인간이 읽을 수 있는 label 값으로 데이터에 접근하는 것이다. 

```python
juice.iloc[0:3, 0:2]
```

```python
juice.loc[0:2, ["Date", "Location"]]
```

#### iloc 조건문
- 조건문을 쓸 때 힘들다.

```python
juice.iloc[juice["Temperature"]==82, 0:2]
```

```python
juice.loc[juice["Temperature"]==82, ["Date", "Location", "Temperature"]]
```

#### 다중 조건식
- data.loc[() & () | (), 열]

### shift()

```python
temp_df = sales[["datesold", "price"]].copy()
temp_df["shifted_v1"] = temp_df["price"].shift(1, fill_value=0).astype(int)
temp_df["shifted_v2"] = temp_df["price"].shift(2, fill_value=0).astype(int)
temp_df
```

```python
temp_df["차이값"] = temp_df["price"] - temp_df["shifted_v1"]
temp_df
```

```python
# 0000년 00월 00일
temp_df["한글날짜"] = temp_df["datesold"].dt.strftime("%Y년 %m월 %d일")
temp_df
```
