### 정규화 작업 (by. SalePrice)
- 정규화 작업 : 전통적인 모델을 활용하기 위해서
 + 전통모델 : 회귀모형 등
- 정규분포를 만드는 이유 : 전통적인 모델의 Assumption
 + 데이터는 정규분포여야 함
 + 정규분포 작업 (log, sqrt 등등)
 + 샤피로, 윌크스 검정 (p-value > 0.5)
 ++ 모수 통계 / 비모수 통계

- 비선형 모델 : 결정트리, 랜덤포레스트, XGBoost, LightGBM
 + 데이터가 정규분포인지 아닌지 상관없음
 + 논리구조가 ***if*** 조건

ex) 500000 이상 데이터에 대해서 loss 가 생김
- 이러한 원인들을 다 모으면 total loss
- 로컬미니마
 + 그래프의 기울기를 미세하게 조정하면서 오차의 총합이 가장 작은 값을 찾아야하는데 가장 작은 값을 찾기도 전에 오차율이 점점 줄어들다가 다시 늘어나는 **"중간지역의 최소값"** 을 전체 함수에서 가장 작은 값으로 착각하는 것
- 이 데이터에 의해서 정확도가 달라질 수 있음
 + 모델을 데이터에 정확도를 맞추면 total loss 정확도가 떨어질 수 있고
 + tatal loss 정확도에 맞추면 데이터의 정확도가 떨어질 수 있음
 + log 변환으로 해결

```python
# 로그변환

import numpy as np

train["SalePrice"] = np.log1p(train["SalePrice"])

(mu, sigma) = norm.fit(train["SalePrice"])
print(mu, sigma)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(train["SalePrice"], color = "b", stat = "probability")
ax.axvline(mu, color="r", linestyle="--")
ax.text(mu + 0.05, 0.105, "Mean of SalePrice", color="r")
plt.show()
```

## 결측치 확인
- 결측치의 비율을 확인하는 사용자 정의 함수
- 결측치 처리: 
 + 데이터 설명서 확인
 + 어떻게 처리할지 판단

```python
def check_na(data, head_num = 6):
  isnull_na = (data.isnull().sum() / len(data)) * 100
  data_na = isnull_na.drop(isnull_na[isnull_na == 0].index).sort_values(ascending=False)
  missing_data = pd.DataFrame({'Missing Ratio' :data_na, 
                               'Data Type': data.dtypes[data_na.index]})
  print("결측치 데이터 컬럼과 건수:\n", missing_data.head(head_num))

check_na(all_df, 20)
```
